````python
# ==============================
# 3. VERIFY FOLDER STRUCTURE
# ==============================

print("Folders inside pulse:")
print(os.listdir("pulse"))
````

````python
# ==============================
# 4. SET DATA PATH
# ==============================

path = "pulse/data/aggregated/insurance/country/india/state"

# Validate path
if not os.path.exists(path):
    raise FileNotFoundError("Path not found. Check clone step.")

print("✅ Path is correct")
print("Sample states:", os.listdir(path)[:5])
````

````python
# ==============================
# 5. DATA EXTRACTION FUNCTION
# ==============================

def extract_data(base_path):
    data_list = []

    try:
        states = os.listdir(base_path)
    except Exception as e:
        raise Exception(f"Error accessing base path: {e}")

    for state in states:
        state_path = os.path.join(base_path, state, "2024")

        if not os.path.isdir(state_path):
            continue

        files = os.listdir(state_path)

        for file in files:
            file_path = os.path.join(state_path, file)

            try:
                with open(file_path, "r") as f:
                    data = json.load(f)

                    for item in data["data"]["transactionData"]:
                        for pi in item["paymentInstruments"]:
                            data_list.append({
                                "state": state,
                                "year": 2024,
                                "quarter": int(file.replace(".json", "")),
                                "count": pi["count"],
                                "amount": pi["amount"]
                            })

            except Exception as e:
                print(f"Skipping file {file_path}: {e}")

    df = pd.DataFrame(data_list)

    if df.empty:
        raise ValueError("No data extracted. Check dataset path.")

    return df
````

````python
# ==============================
# 6. LOAD DATA
# ==============================

df = extract_data(path)

print("✅ Data loaded successfully")
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# ==============================
# DATA WRANGLING
# ==============================

# 1. Copy dataset (best practice)
df_wrangle = df.copy()

# ------------------------------
# 2. Clean State Names
# ------------------------------
df_wrangle['state'] = df_wrangle['state'].str.replace('-', ' ')
df_wrangle['state'] = df_wrangle['state'].str.title()

# ------------------------------
# 3. Handle Missing Values
# ------------------------------
print("Missing values before:\n", df_wrangle.isnull().sum())

# Drop missing values (if any)
df_wrangle = df_wrangle.dropna()

print("Missing values after:\n", df_wrangle.isnull().sum())

# ------------------------------
# 4. Remove Duplicate Records
# ------------------------------
print("Duplicate rows before:", df_wrangle.duplicated().sum())

df_wrangle = df_wrangle.drop_duplicates()

print("Duplicate rows after:", df_wrangle.duplicated().sum())

# ------------------------------
# 5. Data Type Conversion
# ------------------------------
df_wrangle['year'] = df_wrangle['year'].astype(int)
df_wrangle['quarter'] = df_wrangle['quarter'].astype(int)
df_wrangle['count'] = df_wrangle['count'].astype(int)
df_wrangle['amount'] = df_wrangle['amount'].astype(float)

# ------------------------------
# 6. Feature Engineering
# ------------------------------
# Create new feature: Average Transaction Value
df_wrangle['avg_transaction_value'] = df_wrangle['amount'] / df_wrangle['count']

# ------------------------------
# 7. Outlier Detection
# ------------------------------
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.boxplot(x=df_wrangle['amount'])
plt.title("Outlier Detection in Transaction Amount")
plt.show()

# ------------------------------
# 8. Final Check
# ------------------------------
print(df_wrangle.info())
df_wrangle.head()
````

````python
# ==============================
# SQL IMPLEMENTATION (SQLite)
# ==============================

import sqlite3

# Create connection
conn = sqlite3.connect('phonepe.db')

# Store dataframe into SQL table
df_wrangle.to_sql('transactions', conn, if_exists='replace', index=False)

# Query: Top states by transaction amount
query = """
SELECT state, SUM(amount) as total_amount
FROM transactions
GROUP BY state
ORDER BY total_amount DESC
"""

# Run query and display result
result = pd.read_sql(query, conn)

print("Top States by Transaction Amount (SQL Result):")
result.head()
````
